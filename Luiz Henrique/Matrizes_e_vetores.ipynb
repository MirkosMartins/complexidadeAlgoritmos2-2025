```python
import time
import matplotlib.pyplot as plt
import random
```

```python
tamanhos = list(range(1, 501))
tamanhos_matriz = [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000]

```

```python


tempos_vetor = []

def soma_vetores(v1, v2):
    return [v1[i] + v2[i] for i in range(len(v1))]

for N in tamanhos:
    v1 = [random.randint(0, 9) for _ in range(N)]
    v2 = [random.randint(0, 9) for _ in range(N)]

    inicio = time.time_ns()
    soma_vetores(v1, v2)
    fim = time.time_ns()

    tempos_vetor.append((fim - inicio) / 1_000_000)

plt.plot(tamanhos, tempos_vetor, label='Soma Vetores', marker='', color='blue')
plt.xlabel("Tamanho do vetor")
plt.ylabel("Tempo (ms)")
plt.legend()
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt

tempos_vetor = []

def multiplica_vetor_matriz(v, M):
    n = len(v)
    resultado = [0] * n
    for i in range(n):
        for j in range(n):
            resultado[i] += v[j] * M[j][i]
    return resultado

for N in tamanhos:
    v1 = [random.randint(0, 9) for _ in range(N)]
    m1 = [[random.randint(0, 9) for _ in range(N)] for _ in range(N)]

    inicio = time.time_ns()
    multiplica_vetor_matriz(v1, m1)
    fim = time.time_ns()

    tempos_vetor.append((fim - inicio) / 1_000_000)

plt.plot(tamanhos, tempos_vetor, label='Soma Vetores', marker='', color='red')
plt.xlabel("Tamanho do vetor")
plt.ylabel("Tempo (ms)")
plt.legend()
plt.show()
```

```python

tempos_vetor = []

def multiplica_matrizes(A, B):
    linhas_A, colunas_A = len(A), len(A[0])
    colunas_B = len(B[0])
    C = [[0] * colunas_B for _ in range(linhas_A)]

    for i in range(linhas_A):
        for j in range(colunas_B):
            for k in range(colunas_A):
                C[i][j] += A[i][k] * B[k][j]
    return C

for N in tamanhos:
    m2 = [[random.randint(0, 9) for _ in range(N)] for _ in range(N)]
    m3 = [[random.randint(0, 9) for _ in range(N)] for _ in range(N)]

    inicio = time.time_ns()
    multiplica_matrizes(m3, m2)
    fim = time.time_ns()

    tempos_vetor.append((fim - inicio) / 1_000_000)

plt.plot(tamanhos, tempos_vetor, label='Soma Vetores', marker='', color='purple')
plt.xlabel("Tamanho do vetor")
plt.ylabel("Tempo (ms)")
plt.legend()
plt.show()
```
