```python
from random import randint
import time
import matplotlib.pyplot as plt
import numpy as np

# Função para criar lista ordenada de forma decrescente
def criaListaI(numero):
    lista = []
    for i in range(numero):
        lista.append(100-i)
    return lista

# Função para criar lista de números aleatórios entre 1 e 50
def criaLista(numero):
    lista = []
    for i in range(numero):
        lista.append(randint(1,50))
    return lista

# Função do Bubblesort
def bubblesort(lista):
    tamanho = len(lista)
    for i in range(tamanho-1):
        for j in range(tamanho-1):
            if lista[j]>lista[j+1]:
                lista[j], lista[j+1] = lista[j+1], lista[j]
    return lista

# Função do Quicksort
def quicksort(lista):
    if len(lista) <= 1:
        return lista  # Condição de parada: lista já está ordenada
    pivot = lista[0]  # Escolhe o primeiro elemento como pivô
    menores = [x for x in lista[1:] if x <= pivot] # Elementos menores que o pivô
    maiores = [x for x in lista[1:] if x > pivot] # Elementos maiores que o pivô

    # Aplica o quicksort recursivamente nas sublistas e concatena
    return quicksort(menores) + [pivot] + quicksort(maiores)

tempos_bubblesort = []
tempos_quicksort = []

for i in range(100, 1100, 50):  
    l_bubble = criaListaI(i)
    l_quick = criaListaI(i)

    # Tempo para Bubblesort
    inicio = time.time_ns()
    bubblesort(l_bubble)
    fim = time.time_ns()
    tempos_bubblesort.append(fim - inicio)

    # Tempo para Quicksort
    inicio = time.time_ns()
    quicksort(l_quick)
    fim = time.time_ns()
    tempos_quicksort.append(fim - inicio)

# Criando dois gráficos lado a lado
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Bubblesort
ax[0].scatter(range(100, 1100, 50), tempos_bubblesort, color='blue', label='Bubblesort', alpha=0.6)
ax[0].set_xlabel('Número de elementos')
ax[0].set_ylabel('Tempo de execução (ns)')
ax[0].set_title('Bubblesort')
ax[0].legend()

# Quicksort
ax[1].scatter(range(100, 1100, 50), tempos_quicksort, color='red', label='Quicksort', alpha=0.6)
ax[1].set_xlabel('Número de elementos')
ax[1].set_ylabel('Tempo de execução (ns)')
ax[1].set_title('Quicksort')
ax[1].legend()

plt.tight_layout()

plt.show()
```
