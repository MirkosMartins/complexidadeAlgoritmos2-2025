```python
import matplotlib.pyplot as plt
import time
from random import randint

def criaListaI(numero):
    lista=[]
    for i in range(numero):
        lista.append(100-1)
    return lista

def criaLista(numero):
    lista = []
    for i in range(numero):
        lista.append(randint(1,50))
    return lista
```

```python
#QuikSort

def quicksort(arr):
    if len(arr) <= 1: # Caso base: listas com 0 ou 1 elemento já estão ordenadas
        return arr
    else:
        pivo = arr[0] # Escolhe o primeiro elemento como pivô
        menores = [x for x in arr[1:] if x < pivo] # Cria lista de elementos menores que o pivô
        maiores = [x for x in arr[1:] if x >= pivo] # Cria lista de elementos maiores ou iguais ao pivô
        return quicksort(menores) + [pivo] + quicksort(maiores) # Retorna a lista ordenada


tempos = []
for i in range(1000):
    l = criaListaI(1000+i)
    ini = time.time_ns()
    quicksort(l)
    fim = time.time_ns()
    tempos.append(fim-ini)
    plt.scatter(i,tempos[i],c='red')
tempos = []
for i in range(1000):
    l = criaListaI(10+i)
    ini = time.time_ns()
    quicksort(l)
    fim = time.time_ns()
    tempos.append(fim-ini)
    plt.scatter(i,tempos[i],c='black')
plt.title('Quicksort-Lista Invertida')
plt.xlabel('Nr elementos')
plt.ylabel('tempo exec (ns)')
plt.show()
```

```python
# bubblesort

def bubblesort(lista):
    tamanho = len(lista)
    for i in range(tamanho-1):
        for j in range(tamanho-1):
            if lista[j]>lista[j+1]:
                lista[j],lista[j+1]=lista[j+1],lista[j]
    return lista

tempos = []
for i in range(1000):
    l = criaListaI(1000+i)
    ini = time.time_ns()
    bubblesort(l)
    fim = time.time_ns()
    tempos.append(fim-ini)
    plt.scatter(i,tempos[i],c='red')
tempos = []
for i in range(1000):
    l = criaListaI(10+i)
    ini = time.time_ns()
    bubblesort(l)
    fim = time.time_ns()
    tempos.append(fim-ini)
    plt.scatter(i,tempos[i],c='black')
plt.title('Bubblesort-Lista Invertida')
plt.xlabel('Nr elementos')
plt.ylabel('tempo exec (ns)')
plt.show()
```

```python
# Sort do python

tempos = []
for i in range(1000):
    l = criaListaI(1000+i)
    ini = time.time_ns()
    l.sort()
    fim = time.time_ns()
    tempos.append(fim-ini)
    plt.scatter(i,tempos[i],c='red')
tempos = []
for i in range(1000):
    l = criaListaI(10+i)
    ini = time.time_ns()
    l.sort()
    fim = time.time_ns()
    tempos.append(fim-ini)
    plt.scatter(i,tempos[i],c='black')
plt.title('Sort do Python-Lista Invertida')
plt.xlabel('Nr elementos')
plt.ylabel('tempo exec (ns)')
plt.show()
```
